```python
import sys
import numpy as np
import sacc
import yaml
sys.path.insert(1, '/home/jaimerz/Documents/UCL/pacc')
import pacc
```

## DESY1

```python
s = sacc.Sacc.load_fits("/home/jaimerz/Documents/UCL/3x2_analytical/data/HSC/HSC.sacc")
y = yaml.load(open("/home/jaimerz/Documents/UCL/3x2_analytical/data/HSC/HSC.yml"), Loader=yaml.FullLoader)
```

```python
cle = pacc.ClsEnsemble(s,y)
```

```python
wanted_pairs = [['lens_0', 'lens_0'],
                ['lens_1', 'lens_1'],
                ['lens_2', 'lens_2'],
                ['lens_3', 'lens_3']]
```

```python
clse = pacc.ClSubSemble(cle, wanted_pairs)
clse.cov
sig = np.sqrt(np.diag(clse.cov))
corr = clse.cov / np.outer(sig, sig)
```

```python
pacc.plot_cls([cle], wanted_pairs)
```

```python
pacc.plot_cls([cle, cle], cle.pairs[:10])
```

```python
pacc.plot_cov(cle, wanted_pairs, logscale=True)
```

```python
pacc.plot_corr(cle, wanted_pairs, logscale=False)
```
